```python
# prep (already in your notebook - you don't have to reproduce this part)
# btw, if you obtain an error in this cell, just run it again
import matplotlib.pyplot as plt
import numpy as np
# 1000 donn√©es
x = np.random.randn(1000)
y = pd.date_range('1/1/2019', periods=1000)
df = pd.DataFrame(data=x, index=y)
# nous pouvons changer le nom des colonnes
df.columns = ['Valeur']
df = df.cumsum()

# exercice 3
plt.figure(figsize=(9, 9))
plt.title("Pomme Share Price")
plt.xlabel('Date')
plt.ylabel("Share price in USD($)")
plt.plot(df, 'yo')

plt.legend(['POMM'])

plt.show()
```

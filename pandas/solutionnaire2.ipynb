### Exercice 3

```python
# préparation (déja dans votre notebook, vous n'avez pas à repreduire cette section)
# en passant, si vous obtenez une erreur içi, faites juste repartir cette cellule
import matplotlib.pyplot as plt
import numpy as np
# 1000 données
x = np.random.randn(1000)
y = pd.date_range('1/1/2019', periods=1000)
df = pd.DataFrame(data=x, index=y)
# nous pouvons changer le nom des colonnes
df.columns = ['Valeur']
df = df.cumsum()

# exercice 3
plt.figure(figsize=(9, 9))
plt.title("Action de la compagnie Pomme")
plt.xlabel('Évolution dans le temps')
plt.ylabel("Valeur de l'action en USD($)")
plt.plot(df, 'yo')

plt.legend(['POMM'])

plt.show()
```

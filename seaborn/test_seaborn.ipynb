```python
import matplotlib
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
sns.set() # set the matplotlib context and style to seaborn's
dots = sns.load_dataset("tips")

tips = sns.load_dataset("tips")
tips.head(1)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="sex", size="size", data=tips)
```

```python
dots = sns.load_dataset("dots")
dots.head(1)
```

```python
sns.relplot(x="time", y="firing_rate", col="align", hue="choice", size="coherence",
            facet_kws=dict(sharex=True), kind="line", data=dots)

```

```python
fmri = sns.load_dataset("fmri")
fmri13 = fmri[fmri.subject == "s13"]
fmri.head(4)
```

```python
sns.relplot(x="timepoint", y="signal", col="region", hue="event", style="event",kind="line",data=fmri)

```

```python
sns.lmplot(x="total_bill", y="tip", col="time", hue="smoker",x_estimator=lambda x: 0.15 *, data=tips)
 
```

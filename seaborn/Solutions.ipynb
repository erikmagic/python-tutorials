### Exercise !
> Create a `relplot` using the single_fmri data with a white background, no spines on the sides and dark grey grid lines


```python
fmri_data2 = sns.load_dataset("fmri")
single_fmri = fmri_data2[fmri_data2.subject=="s0"].drop(['event', 'region'], axis=1)

with sns.axes_style("whitegrid", {"axes.spines.top": False, "axes.spines.right": False, "grid.color": "darkgrey"}):
    g = sns.relplot(x="timepoint", y="signal", data=single_fmri)
    g.ax.set_title('ticks')
```

```python
titanic_df = sns.load_dataset("titanic")
titanic_df = titanic_df.dropna(subset=['age'])
len(titanic_df)

sns.set_palette("BuGn_r")
sns.set_style("white")
ax = sns.distplot(titanic_df.fare, bins=4, kde=False)
ax.set_title("Histogram of the Fares Paid by Titanic's Customers")
ax.set_xlim(-10,300)
ax.set_ylabel("Fare paid ($)")
```

### Exercise!
 Reproduce the following graph using the pokemon dataset.

```python
pokemon_dataset = pd.read_csv("Pokemon.csv")
pokemon_dataset.head()

sns.set_style('ticks')
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16,16))

speed_attack = sns.kdeplot(pokemon_dataset.Attack, pokemon_dataset.Speed, ax = ax[0][0], shade=True, shade_lowest=False, cmap="Greens")
speed_attack.set_title("Speed vs Attack")

defense_attack = sns.kdeplot(pokemon_dataset.Attack, pokemon_dataset.Defense, ax =ax[0][1], shade=True, shade_lowest=False, cmap="Blues")
defense_attack.set_title("Defense vs Attack")

hp_attack = sns.kdeplot(pokemon_dataset.Attack, pokemon_dataset.HP, ax =ax[1][0], shade=True, shade_lowest=False, cmap="Reds")
hp_attack.set_title("HP vs Attack")

speed_speed = sns.kdeplot(pokemon_dataset['Sp. Atk'], pokemon_dataset['Sp. Def'], ax =ax[1][1], shade=True, shade_lowest=False, cmap="Oranges")
speed_speed.set_title("Defense Speed vs Attack Speed")


```
